```python
gp1 = storagecosts.groupby('productcode').agg({'QTYScrap':'sum','VALScrap':'sum'}).reset_index().rename(columns={'QTYScrap':'Sum_Qtyscrap', 'VALScrap':'Sum_VALScrap'})
gp1.sort_values(['Sum_Qtyscrap'],ascending=False).head(50)
```

```python
new_df = pd.merge(storagecosts,gp1, on='productcode', how='left')
new_df.columns
```

```python
new_df["source"] = np.where(new_df["source"] == "sealy","Sealy",new_df["source"])
new_df[new_df['Sum_Qtyscrap']>150]['source'].value_counts(normalize='true')
import seaborn as sns
sns.barplot(x='source',y='Sum_Qtyscrap',data=new_df[new_df['Sum_Qtyscrap']>150])
```
